```python
import pandas as pd    
import gc
import os

review = pd.read_excel(os.path.join('..','..','data','JS_11_data_review.xlsx'))
print(review.shape)
review.head()
```

```python
dd = review.groupby('id')['rating'].agg([('ids','mean')]).reset_index()
dd = dd.reset_index()
dd = dd.rename(columns = {'index':'userID'})
ds = pd.merge(review,dd, on='id')
ds = ds.drop(columns = ['ids','id','language','idno','Unnamed: 0','body'], axis=1)
review = ds.rename(columns = {'place.id':'locationId'})

d1 = review.groupby('locationId')['photonum'].agg([('average_photonum','mean')]).reset_index()
review = pd.merge(review,d1, on='locationId')
d2 = review.groupby('locationId')['rating'].agg([('average_rating','mean')]).reset_index()
review = pd.merge(review,d2, on='locationId')
d4 = review.groupby('userID')['rating'].agg([('user_mean_rating','mean')]).reset_index()
review = pd.merge(review,d4, on='userID')
d5 = review.groupby('userID')['rating'].agg([('user_reviewcount','count')]).reset_index()
review = pd.merge(review,d5, on='userID')
review
```

```python
place = place.rename(columns = {'place.id':'locationId'})
place = place.drop(columns = ['Unnamed: 0','place.category','place.name'])
place = place.rename(columns = {'review.count':'rated_count'})
place['category_l'] = place['category'].apply(lambda x: 1 if x=='EAT' else 0)
place = place.drop(columns = ['is_fch'])
place = place.drop(columns = ['rating'])
place
```

```python
df = pd.merge(review, place, on='locationId')
print(df.shape)
df
```

```python
df.to_csv(os.path.join("..","..","data","YN_final_df2.csv"), index=False)
print('save the data YN_final_df2')
```
